```python
#Imports
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
%matplotlib inline
```

```python
#use datareader to grab info on the stock from Jan 1st 2006 to Jan 1st 2016 for each of these banks.

# Bank of America
BAC = data.DataReader("BAC", 'yahoo', start='2006-01-01', end='2016-01-01' )
# City Group
C = data.DataReader("C", 'yahoo', start='2006-01-01', end='2016-01-01' )
# Goldman Sachs
GS = data.DataReader("GS", 'yahoo', start='2006-01-01', end='2016-01-01' )
# JPMorgan Chase
JPM = data.DataReader("JPM", 'yahoo', start='2006-01-01', end='2016-01-01' )
# Morgan Stanley
MS = data.DataReader("MS", 'yahoo', start='2006-01-01', end='2016-01-01' )
# Wells Fargo
WFC= data.DataReader("WFC", 'yahoo', start='2006-01-01', end='2016-01-01' )
```

```python
#Creating tickers list of the ticker symbols (as strings) in alphabetical order.

tickers = ['BAC','C','GS','JPM','MS','WFC']
```

```python
#concatenate the bank dataframes together to a single data frame called bank_stocks.

bank_stocks = pd.concat([BAC,C,GS,JPM,MS,WFC], axis =1, keys=tickers)
bank_stocks.head()
```

```python
#Set the column name levels 
bank_stocks.columns.names = ['Bank Ticker','Stock Info']

#Check the head of the bank_stocks dataframe
bank_stocks.head(5)
```

```python
#EDA
#Let's explore the data a bit!
```

```python
# What is the max Close price for each bank's stock throughout the time period?
bank_stocks.xs(key='Close', axis=1, level= 'Stock Info' ).max()
```

```python
#Create a new empty DataFrame called returns.
returns = pd.DataFrame()
```

```python
#We can use pandas pct_change() method on the Close column to create a column representing this return value. 
#Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in
#the returns DataFrame.

for x in tickers:
    returns[x + '  Returns']= bank_stocks[x]['Close'].pct_change()
returns.head()
```

```python
#Create a pairplot for exploring data using seaborn of the returns dataframe.

import seaborn as sns
sns.pairplot(returns[1:])
```

```python
#on what dates each bank stock had the best and worst single day returns.
```

```python
#Take a look at the standard deviation of the returns, 
#which stock is the riskiest over the entire time period?

returns.std()
```

```python
#distplot of the 2015 returns for Morgan Stanley

sns.distplot(returns.loc['2015-01-01':'2015-12-31']['MS  Returns'], color='green', bins=100)
```

```python
#imports

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline

#Plotly Method Imports
import plotly
import cufflinks as cf
cf.go_offline()
```

```python
#line plot showing Close price for each bank for the entire index of time.

bank_stocks.xs(key='Close', axis=1, level= 'Stock Info').plot(figsize = (10,6))
```

```python
#Interactive plot using plotly

bank_stocks.xs(key='Close', axis=1, level= 'Stock Info').iplot()
```

```python
#analyze the moving averages for these stocks in the year 2008.
#Plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008.

BAC['Close'].loc['2008-01-01':'2009-01-01'].rolling(window=10).mean().plot(figsize=(12,6),color='green', label='30 days avg')
BAC['Close'].loc['2008-01-01':'2009-01-01'].plot(label= 'BAC Close')
plt.legend()
plt.tight_layout()
```

```python
#heatmap of the correlation between the stocks Close Price.

sns.heatmap(bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr(),annot=True)
```

```python
#clustermap to cluster the correlations together:

sns.clustermap(bank_stocks.xs(key='Close',axis=1, level='Stock Info').corr(), annot= True)
```

```python
#interactive heatmap

close_corr = bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr()
close_corr.iplot(kind= 'heatmap')
```
